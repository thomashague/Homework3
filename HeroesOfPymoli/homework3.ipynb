```python
import pandas as pd
import numpy as np
```

```python
#get path (remember to make this adjustable based on file)
path = "../HeroesOfPymoli/purchase_data.json"
file = pd.read_json(path)

file.head()
```

```python
#get the number of unique players (unique screen names in SN column)
numPlayers = file.SN.nunique()
print("There are " + str(numPlayers) + " players")
#unique items
uniqueItems = file['Item Name'].nunique()
print("There are " + str(uniqueItems) + " unique Items")
#average price
avgPrice = np.mean(file.Price)
print("There average price of items is " + str(numPlayers))
#total purchases
totalPurchases = file.Price.count()
print("There was a total of " + str(totalPurchases) + " purchases")
#total revenue
totalRevenue = file["Price"].sum()//1
print("The total revenue is " + str(totalRevenue))
```

```python
uniqueSN = file.groupby(["SN", "Gender"],as_index=False).count()


counts = uniqueSN['Gender'].value_counts().to_dict()
totalMale = counts["Male"]//1
totalFemale = counts["Female"]//1
totalOther = numPlayers - (totalMale + totalFemale)
percMale = ((totalMale/(totalMale + totalFemale + totalOther))*100)//1
percFemale = ((totalFemale/(totalMale + totalFemale + totalOther))*100)//1
percOther = ((totalOther/(totalMale + totalFemale + totalOther))*100)//1

print(str(totalMale) + " (" + str(percMale) + "%) players are Male")
print(str(totalFemale) + " (" + str(percFemale) + "%) players are Female")
print(str(totalOther) + " (" + str(percOther) + "%) players are Other / Non-Disclosed")
```

```python
#uniqueSN = file.groupby("Gender",as_index=True).count()
maleFile = file.loc[file['Gender'] == 'Male']
femaleFile = file.loc[file['Gender'] == 'Female']
otherFile = file.loc[file['Gender'] == 'Other / Non-Disclosed']

malePurchCount = maleFile.Price.count()
femalePurchCount = femaleFile.Price.count()
otherPurchCount = otherFile.Price.count()

avgMalePrice = np.mean(maleFile.Price)
avgFemalePrice = np.mean(femaleFile.Price)
avgOtherPrice = np.mean(otherFile.Price)

sumMaleValue = np.sum(maleFile.Price)
sumFemaleValue = np.sum(femaleFile.Price)
sumOtherValue = np.sum(otherFile.Price)

print("Males bought " + str(malePurchCount) + " items at an average price of $" + str(avgMalePrice))
print("Females bought " + str(femalePurchCount) + " items at an average price of $" + str(avgFemalePrice))
print("The total purchase value for Males is " + str(sumMaleValue))
print("The total purchase value for Females is " + str(sumFemaleValue))

#normalized totals????????
```
